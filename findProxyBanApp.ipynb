Normally when testing app that ban proxy, the time taken to generate all actions are very low --> less than 1 minute

```python
import os
import re
from datetime import datetime, timedelta
```

```python
def checkTestPeriod(contents):
    time_regex = re.compile(r'\d{2}:\d{2}:\d{2}\.\d{3}')
    # search for the time in the string
    lines = contents.splitlines()
    startTime = 0
    endTime = 0
    # loop through each line and search for the line with "Action count: 300"
    for line in lines:
        if "Action count: 300" in line:
            match = time_regex.search(line)
            if match:
                time_str = match.group(0)
                startTime = time_str
        if "The application has been tested sucessfully" in line:
            match = time_regex.search(line)
            if match:
                time_str = match.group(0)
                endTime = time_str
    if startTime != 0 and endTime != 0:
        start_time = datetime.strptime(startTime, "%H:%M:%S.%f")
        end_time = datetime.strptime(endTime, "%H:%M:%S.%f")
        time_diff = end_time - start_time
        if time_diff < timedelta(minutes=1):
            return False
        else:
            return True
        
    
```

```python
# define the path to the folder containing the text files
folder_path = "./log_tester"
noTest = {} # appId: 1
testTimeOut = {} # start test but timeout
testSuccess = {}
proxyBan = {}


# loop through all the files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    if os.path.isfile(file_path):

        # read the contents of the file
        with open(file_path, 'r') as f:
            try:
                contents = f.read()
            except:
                print(filename, 'fail to read log')

        # check if the contents contain at least one sentence
        if re.search('Action count: 300', contents):
            if re.search('The application has been tested sucessfully', contents):
                if checkTestPeriod(contents):
                    testSuccess[filename] = 1
                    print(filename, 'is success')
                else:
                    proxyBan[filename] = 1
                    print(filename, 'is proxy ban')
                
            else:
                testTimeOut[filename] = 1
                print(filename, 'test time out = proxy not ban')
        else:
            noTest[filename] = 1
            print(filename, 'no test')
```

```python
toFind = '200 OK'
folder_path = "./log_mitm"
hasProxy = {}
noProxy = {}


# loop through all the files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    if os.path.isfile(file_path):

        # read the contents of the file
        with open(file_path, 'r') as f:
            try:
                contents = f.read()
            except:
                print(filename, 'fail to read log')

        # check if the contents contain at least one sentence
        if re.search(toFind, contents):
            hasProxy[filename] = 1
                
        else:
            noProxy[filename] = 1
```
